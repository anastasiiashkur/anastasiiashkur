Data Cleaning & Manipulation

```python
# loading libraries
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1L4aScG_IYBOnCCgtjJ1yLKUhi6n68YGP')
df
```

```python
# Calculate the 25th percentile.
qOne = df['AGE'].quantile(0.25)

# Calculate the 75th percentile.
qThree = df['AGE'].quantile(0.75)

# Calculate Interquartile Range
iqr = stats.iqr(df['AGE'])

# "Maximum" and "Minimum"
maximum = qThree + (1.5 * iqr)
minimum = qOne - (1.5 * iqr)

# Exact outlier values
iqrOutliers = df['AGE'].loc[(df['AGE'] < minimum) | (df['AGE'] > maximum)]

# Remove outlier values
df['AGE'].drop(iqrOutliers.index)
```

```python
# Remove outlier values from dataset
df_wihhout_outlets = df.drop(iqrOutliers.index)
df_wihhout_outlets
```

```python
# str in numeric
df_wihhout_outlets['HYPERTEN_int'] = df_wihhout_outlets['HYPERTEN'].apply(lambda x: 1 if x == '1.0' else 0)
df_wihhout_outlets
```
